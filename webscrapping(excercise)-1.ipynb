```python
from bs4 import BeautifulSoup
import requests
```

```python
url ="https://realpython.com/"
response = requests.get(url)
response.raise_for_status()
```

```python
soup = BeautifulSoup(response.content, "html.parser")
print(soup.prettify())
```

```python
titles = soup.find_all("h2", class_="card-title")
titles
```

```python
data = []
for i in titles:
    print(i.text)
    data.append(i)
data
```

```python
Dic = {"Card-Titles": data}
Dic
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data, columns=['Card-Titles']) #Created the DataFrame directly
df.index = df.index + 1 #Added 1 to each index value

df


```
